# Data Clean

I decide to drop the column url because it is not helpful for this analysis

```python
ted = ted.drop(columns=['url', 'related_talks'])
```

```python
ted.head()
```

I will convert the "film_date" and "published_date" into human readable format

```python
import datetime
ted['film_date'] = ted['film_date'].apply(lambda x: datetime.datetime.fromtimestamp( int(x)).strftime('%d-%m-%Y'))
ted['published_date'] = ted['published_date'].apply(lambda x: datetime.datetime.fromtimestamp( int(x)).strftime('%d-%m-%Y'))
```

```python
ted.head()
```

I will convert the "tags" values into columns in one-hot encoding type

```python
def list2columns(df, columns):
    """
    to convert list in the columns 
    of a dataframe
    """
    for col in columns:    
        for i in range(len(df[col])):
            # get the list of columns to set
            for element in eval(df.loc[i,col]):
                if len(element)!=0:
                    if element not in df.columns:
                        df.loc[:,element]= 0
                    df.loc[i,element] = 1
            if i >= 6:
                if df.loc[6, 'Christianity'] == 0:
                    print(i)
                    break
    return df
```

```python
ted = list2columns(ted, ['tags'])
print(ted.loc[0, 'children'])
```
